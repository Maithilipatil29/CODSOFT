```python
#Histplot of survived people

sns.set(style="white")
sns.histplot(df["Survived"], kde=True,bins=10)
```

```python
#Heatmaps for dataset

tc=df.corr()
sns.heatmap(tc)
```

```python
#Heatmap of the correlated matrix for dataset

sns.heatmap(tc, annot=True, cmap="plasma", linecolor="black", linewidths=1)
```

```python
#Histplot of all the dataset

tc=df.corr()
sns.set(style="dark")
sns.histplot(tc, kde=True, bins=10)
```

```python
#Countplot of Survived or not survived as having value 0 and 1

sns.countplot(x=df['Survived'],data=df)
```

```python
#Barplot: Sex vs Survived

sns.set_style('darkgrid')

sns.barplot(x=df['Sex'],y=df['Survived'], data=df)
```

```python
#Barplot of SibSp vs Survived

sns.set_style('darkgrid')

sns.barplot(x=df['SibSp'],y=df['Survived'], data=df)
```

```python
#Swarmplot of Survived and not not survived persons to their Age

sns.swarmplot(x=df['Survived'], y=df['Age'], data=df)
```
