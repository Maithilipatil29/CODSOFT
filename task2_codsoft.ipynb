```python
#pairplot

sns.pairplot(df, hue="species")
plt.title("species")
plt.show()
```

```python

sns.scatterplot(x='sepal_length',y='sepal_width',data=df,hue='species')

plt.show()
```

```python
#linear regression

df_binary=df[['sepal_length','sepal_width','petal_length','petal_width']]
df_binary.columns=['sepal_length','sepal_width','petal_length','petal_width']
sns.lmplot(x='sepal_length',y='sepal_width', data=df_binary,order=2,ci=None)
sns.lmplot(x='sepal_length',y='petal_width', data=df_binary,order=2,ci=None)
sns.lmplot(x='sepal_length',y='petal_length', data=df_binary,order=2)
plt.show()
```
